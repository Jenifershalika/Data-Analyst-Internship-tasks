```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("marketing_AB.csv")  # use your exact file name
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df['test group'].value_counts()
```

```python
df['converted'] = df['converted'].astype(int)
```

```python
control = df[df['test group'] == 'control']['converted']
test = df[df['test group'] == 'test']['converted']
```

```python
control = df[df['test group'] == 'psa']['converted']
test = df[df['test group'] == 'ad']['converted']
```

```python
# H0: Conversion rate of control and test groups are equal
# H1: Conversion rate of control and test groups are different
# Alpha = 0.05
```

```python
from scipy import stats

t_stat, p_value = stats.ttest_ind(control, test)
p_value
```

```python
if p_value < 0.05:
    print("Statistically significant difference between groups")
else:
    print("No statistically significant difference between groups")
```

```python
summary = pd.DataFrame({
    'Group': ['Control', 'Test'],
    'Users Count': [len(control), len(test)],
    'Conversion Rate': [control.mean(), test.mean()]
})

summary
```

```python
summary.to_csv("ab_test_summary.csv", index=False)
```

```python
with open("final_recommendation.txt", "w") as f:
    f.write(
        "The A/B test was conducted at 95% confidence level. "
        "Based on the p-value, the difference between control and test "
        "groups was evaluated and a data-driven decision can be made."
    )
```
