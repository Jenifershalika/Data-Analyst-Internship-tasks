```python
import os

os.makedirs("raw", exist_ok=True)
os.makedirs("processed", exist_ok=True)
os.makedirs("output", exist_ok=True)
```

```python
import pandas as pd

df = pd.read_csv("Customer_Churn.csv")
df.head()
```

```python
df = df.dropna()
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df.columns
```

```python
df['total_revenue'] = df['monthlycharges'] * df['tenure']
```

```python
df['high_value_customer'] = df['total_revenue'] > 5000
```

```python
customers = df[['customerid','gender','seniorcitizen','partner','dependents']]

services = df[['customerid','phoneservice','internetservice','streamingtv','streamingmovies']]

billing = df[['customerid','monthlycharges','totalcharges','paymentmethod']]

churn = df[['customerid','churn']]
```

```python
customers.to_csv("output/customers.csv", index=False)
services.to_csv("output/services.csv", index=False)
billing.to_csv("output/billing.csv", index=False)
churn.to_csv("output/churn.csv", index=False)
```

```python
import sqlite3

conn = sqlite3.connect("database.sqlite")

customers.to_sql("customers", conn, if_exists="replace", index=False)
services.to_sql("services", conn, if_exists="replace", index=False)
billing.to_sql("billing", conn, if_exists="replace", index=False)
churn.to_sql("churn", conn, if_exists="replace", index=False)

conn.close()
```
